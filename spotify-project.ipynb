## 7. [Follow-up] Can you make this plot more readable by removing outliers (extreme values)?

```python
df[(df['duration_ms']/60000) > 16.67].count() # returns 470 songs to be removed as indicated
dist_df = df[(df['duration_ms']/60000) <= 16.67]

sns.distplot(dist_df['duration_ms']/60000) # Songs in minutes
```

## 15. Can you plot a line graph representing the annual number of songs since 1921?

```python
import matplotlib.pyplot as plt

df_1921 = df.query('year >= 1921' ).groupby("year") ["id"].count().reset_index().sort_values("year", ascending=True) 

ax = sns.lineplot(data=df_1921, x="year", y="id")
ax.set(xlabel='Years', ylabel='Number of Songs')

plt.show()
```
